```python
import pandas as pd
```

```python
merged_1 = pd.read_csv('../US_Datasets/merged_us.csv')
merged_1
```

```python
import numpy as np

# Replace placeholder missing values (e.g., '..') with np.nan
merged_1.replace({'..': np.nan}, inplace=True)
```

```python
# Fill missing values in `gov_effectiveness_perc_rank` and `political_stability_perc_rank`
first_gov_effectiveness = merged_1['gov_effectiveness_perc_rank'].dropna().iloc[0]
first_political_stability = merged_1['political_stability_perc_rank'].dropna().iloc[0]

merged_1['gov_effectiveness_perc_rank'].fillna(first_gov_effectiveness, inplace=True)
merged_1['political_stability_perc_rank'].fillna(first_political_stability, inplace=True)

# Save the updated DataFrame if necessary
merged_1.to_csv("../US_Datasets/merged_us.csv", index=False)
```

```python
merged_1
```

```python
#lagging
```

```python
# Identify columns to lag (all columns except 'Bond Rating' and 'quarter')
columns_to_lag = [col for col in merged_1.columns if col not in ['Bond Rating', 'quarter']]

# Create a new DataFrame to store the lagged features
lagged_df = merged_1[['quarter', 'Bond Rating']].copy()  # Keep 'quarter' and 'Bond Rating' in the new DataFrame

# Add lagged features to the new DataFrame
for col in columns_to_lag:
    lagged_df[f"{col}_lag1q"] = merged_1[col].shift(1)
```

```python
lagged_df
```

```python
# Encode 'Bond Rating' using the extended rating mapping
rating_mapping = {
    'AAA': 1, 'AA+': 2, 'AA': 3, 'AA-': 4,
    'A+': 5, 'A': 6, 'A-': 7,
    'BBB+': 8, 'BBB': 9, 'BBB-': 10,
    'BB+': 11, 'BB': 12, 'BB-': 13,
    'B+': 14, 'B': 15, 'B-': 16,
    'CCC+': 17, 'CCC': 18, 'CCC-': 19
}

# Map the 'Bond Rating' column to numeric values
lagged_df['Bond Rating Encoded'] = lagged_df['Bond Rating'].map(rating_mapping)

# Exclude non-numeric columns
numeric_columns = lagged_df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Extract correlations of all features with the target ('Bond Rating Encoded'), excluding itself
correlation_with_target = correlation_matrix['Bond Rating Encoded'].drop('Bond Rating Encoded')

# Calculate absolute values and sort
absolute_correlation_with_target = correlation_with_target.abs().sort_values(ascending=False)

# Display the absolute correlations
print(absolute_correlation_with_target)
```

```python
data = lagged_df.copy()
```

```python
# Shift target to predict the next quarter's bond rating
data['Bond Rating Target'] = data['Bond Rating Encoded'].shift(-1)

# Drop last row (since it has no target value)
data = data.dropna(subset=['Bond Rating Target'])
```

```python
data.drop(columns = 'Bond Rating Encoded', inplace = True )
```

```python
data
```

```python
# Function to convert 'QX-YYYY' format to datetime
def convert_quarter_to_datetime(quarter_str):
    quarter_map = {'Q1': '-01', 'Q2': '-04', 'Q3': '-07', 'Q4': '-10'}  # Approximate months
    qtr, year = quarter_str.split('-')  # Split into Quarter and Year
    return pd.to_datetime(year + quarter_map[qtr])  # Convert to datetime

# Apply conversion to dataset
data['quarter_dt'] = data['quarter'].apply(convert_quarter_to_datetime)
```

```python
data
```

```python
# Define all features (EXCLUDING 'Bond Rating Encoded')
features = [col for col in data.columns if col not in ['quarter_dt', 'Bond Rating Target', 'quarter']]
```

```python
features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Separate features (X) and target (y)
X = data[features]
y = data['Bond Rating Target']

# **Time-Based Train-Test Split**
train_cutoff_quarter = pd.to_datetime('2021-12-31')

X_train = X[data['quarter_dt'] <= train_cutoff_quarter]
y_train = y[data['quarter_dt'] <= train_cutoff_quarter]
X_test = X[data['quarter_dt'] > train_cutoff_quarter]
y_test = y[data['quarter_dt'] > train_cutoff_quarter]
```

```python
# Scale Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#base model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
