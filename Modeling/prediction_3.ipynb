## Preprocessing

lagging all variables and the target

```python
# separating the variables first and greating just one lag
columns_to_lag = [col for col in merged.columns if col not in ['Bond Rating', 'quarter']]

# saving it in a new dataframe 
lagged_df = merged[['quarter', 'Bond Rating']].copy()

# Add lagged features to the new DataFrame
for col in columns_to_lag:
    lagged_df[f"{col}_lag1q"] = merged[col].shift(1)
```

filling in the nulls

```python
lagged_df.fillna(method='ffill', inplace=True)
lagged_df.fillna(method='bfill', inplace=True)
```

```python
lagged_df
```

Changing quarters to datetime for a time-based split

```python
# Function to convert 'QX-YYYY' format to datetime
def convert_quarter_to_datetime(quarter_str):
    quarter_map = {'Q1': '-01', 'Q2': '-04', 'Q3': '-07', 'Q4': '-10'}  # Approximate months
    qtr, year = quarter_str.split('-')  # Split into Quarter and Year
    return pd.to_datetime(year + quarter_map[qtr])  # Convert to datetime

# Apply conversion to dataset
lagged_df['quarter_dt'] = lagged_df['quarter'].apply(convert_quarter_to_datetime)
```

```python
#renaming the dataframe for easy access
data = lagged_df.copy()
```

```python
# now that we have a datetime function lets split based on time
# Define all features
features = [col for col in data.columns if col not in ['quarter_dt', 'Bond Rating', 'quarter']]
# Separate features (X) and target (y)
X = data[features]
y = data['Bond Rating']

# **Time-Based Train-Test Split**
train_cutoff_quarter = pd.to_datetime('2021-12-31')

X_train = X[data['quarter_dt'] <= train_cutoff_quarter]
y_train = y[data['quarter_dt'] <= train_cutoff_quarter]
X_test = X[data['quarter_dt'] > train_cutoff_quarter]
y_test = y[data['quarter_dt'] > train_cutoff_quarter]
```
