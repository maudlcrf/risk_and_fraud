```python
#Identify the categorical columns
categorical_cols = ["Series Name", "Country Name"]

# Identify the year columns dynamically
year_cols = [col for col in df.columns if col.isdigit() and len(col) == 4]

# Melt the DataFrame to convert years from columns to rows
df_melted = df.melt(id_vars=categorical_cols, var_name='Year', value_name='Value')

# Remove missing values (if needed)
df_melted = df_melted[df_melted["Value"].notna()]

# Create new rows for each quarter
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

# Repeat each row 4 times (one for each quarter)
df_expanded = df_melted.loc[df_melted.index.repeat(4)].reset_index(drop=True)

# Assign quarterly labels
df_expanded['Quarter'] = quarters * (len(df_melted))

# Ensure the correct column order
df_expanded = df_expanded[categorical_cols + ['Year', 'Quarter', 'Value']]
```

```python
df_political_stability = df_expanded[df_expanded["Series Name"].str.contains("Political Stability", na=False)].reset_index(drop=True)
```

```python
df_political_stability.drop(columns = 'Series Name', inplace = True)
```

```python
df_political_stability['quarter'] = df_political_stability['Quarter'].astype(str) + '-' + df_political_stability['Year'].astype(str)
```

```python
df_political_stability.drop(columns = ["Year", "Quarter"], inplace = True)
```
